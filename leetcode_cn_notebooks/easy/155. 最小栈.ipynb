```python
# 设计一个支持 push，pop，top 操作，并能在常量时间内检索最小元素的栈。

# push(x) -- 将元素x推入栈中。
# pop() -- 删除栈顶的元素。
# top() -- 获取栈顶元素。
# getMin() -- 检索栈中的最小元素。
# 示例:

# MinStack minStack = new MinStack();
# minStack.push(-2);
# minStack.push(0);
# minStack.push(-3);
# minStack.getMin();   --> 返回 -3.
# minStack.pop();
# minStack.top();      --> 返回 0.
# minStack.getMin();   --> 返回 -2.
```

```python
class MinStack:

    def __init__(self):
        """
        initialize your data structure here.
        """
        self.data = []
        self.min_data = None
        

    def push(self, x):
        """
        :type x: int
        :rtype: void
        """
        self.data.append(x)
        if self.min_data is None:
            self.min_data = x
        if x < self.min_data:
            self.min_data = x
        

    def pop(self):
        """
        :rtype: void
        """
        res = self.data.pop()
        if res == self.min_data:
            if len(self.data) == 0:
                self.min_data = None
            else:
                self.min_data = min(self.data)
        

    def top(self):
        """
        :rtype: int
        """
        return self.data[-1]
        

    def getMin(self):
        """
        :rtype: int
        """
        return self.min_data


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(x)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

```python
# solution 2

class MinStack:

    def __init__(self):
        """
        initialize your data structure here.
        """
        self.data = []
        self.min_data = []
        

    def push(self, x):
        """
        :type x: int
        :rtype: void
        """
        self.data.append(x)
        if len(self.min_data) == 0:  # 如果 min_data 为空 则直接压入
            self.min_data.append(x)
        else:
            curr_min = self.min_data[-1]
            if x <= curr_min:        # 如果 x 小于等于 当前最小值 也压入
                self.min_data.append(x)

        
    def pop(self):
        """
        :rtype: void
        """
        res = self.data.pop()
        if res == self.min_data[-1]:  # 如果 pop 等于当前最小值 也需要更新 min_data
            self.min_data.pop()


    def top(self):
        """
        :rtype: int
        """
        return self.data[-1]
        

    def getMin(self):
        """
        :rtype: int
        """
        return self.min_data[-1]


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(x)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```
